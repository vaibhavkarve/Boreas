```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing(use_unicode=True)
%matplotlib notebook
```

```python
n = sp.symbols('n', integer=True, positive = True)
k = sp.symbols('k', integer=True, nonnegative=True)
T = sp.symbols('T', real=True)
```

```python
def a(n,k): return (-1)**k*n*T**(n-2*k)*sp.binomial(n-k,k)
def b(n): return sum([a(n,k) for k in range(0, n/2+1, 1)])
def c(n): return sum([b(m) for m in range(1, n+1, 1)])/float(n)
def d(n): return sum([c(m) for m in range(1, n+1, 1)])/float(n)
```

```python
points = np.arange(-2,2, 0.001)
for n in range(1,20):
    plt.plot(points, 0*points, 'k--')
    plt.plot(points, sp.lambdify(T, b(n))(points))
plt.show()
```

```python
points = np.arange(-2,2, 0.001)
for n in range(1,20):
    plt.plot(points, 0*points, 'k--')
    plt.plot(points, sp.lambdify(T, c(n))(points))
plt.show()
```

```python
points = np.arange(-2,2, 0.001)
for n in range(1,20):
    plt.plot(points, 0*points, 'k--')
    plt.plot(points, sp.lambdify(T, d(n))(points))
plt.show()
```
